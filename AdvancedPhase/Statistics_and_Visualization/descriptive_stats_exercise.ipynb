```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
exercise = sns.load_dataset('exercise')
exercise
```

```python
print(exercise['diet'].unique(), '<---DIETS')
print(' ')
print(exercise['kind'].unique(), '<---ACTIVITIES')
print(' ')
print(exercise['time'].unique(), '<---TIMES')
```

```python
# print(exercise['diet'].unique())
# print(exercise['pulse'])

diet_pulse = pd.DataFrame(exercise, columns=['diet', 'pulse', 'kind'])
print(diet_pulse[:30])
```

```python
sns.displot(exercise, x='pulse', kind='kde', color='black')
# sns.set(rc={'axes.facecolor':'black'})
sns.set(style='dark')

# plt.yticks([])
# plt.ylabel('')

mean = exercise.pulse.mean()
median = exercise.pulse.median()
mode = exercise.pulse.mode()[0]

plt.axvline(x=mean, color='blue')
plt.axvline(x=median, color='red')
plt.axvline(x=mode, color='orange')

plt.legend(['Pulse', 'Mean', 'Median', 'Mode'])

```

```python
sns.displot(exercise, x='pulse', hue='time', kind='kde', multiple='stack')
# sns.set(rc={'axes.facecolor':'black'})
sns.set(style='dark')
```

```python
sns.displot(exercise, x='pulse', hue='diet', kind='kde', multiple='fill')
# sns.set(rc={'axes.facecolor':'black'})
sns.set(style='dark')
```

```python
sns.kdeplot(exercise, x='pulse', hue='time', multiple='stack')
# sns.set(rc={'axes.facecolor':'black'})
sns.set(style='dark')
```

```python
sns.histplot(exercise, x='pulse', hue='diet', multiple='stack')
# sns.set(rc={'axes.facecolor':'black'})
sns.set(style='dark')
```

```python
sns.displot(exercise, x='pulse', kind='hist')
# sns.set(rc={'axes.facecolor':'black'})
sns.set(style='dark')
```
